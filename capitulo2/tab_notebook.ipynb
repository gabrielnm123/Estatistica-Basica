```python
# impotando a classe Tabela, que manipula data frames do pandas, do módulo tabela
from tabela.tabela import Tabela

# 1 indica que a tabela é em excel, os demais parametros são os que iriam em pd.read_excel
tabela = Tabela('../Dados_EB.xlsx', sheet_name='Tabela 2.1', header=[1], sort_values_by_variavel='Salario (x Sal Min)')
```

```python
# Tabela 2.1 - tabela original
tabela_2_1 = tabela.tab
tabela_2_1
```

```python
# Tabela 2.2 - tabela de frequência e proporção para variável categórica 'Grau de Instrução'
tabela_2_2 = tabela.frequencia_proporcao('Grau de Instrução')
tabela_2_2
```

```python
# Tabela 2.4 - tabela de frequência para variável contínua 'Salario (x Sal Min)' com amplitude 4
tabela_2_4 = tabela.frequencia_variavel_continua('Salario (x Sal Min)', 4)
tabela_2_4
```
