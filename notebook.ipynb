```python
# Capitulo 2

## Resumo de Dados:

# 2.1 Tipos de variáveis  

import pandas as pd

tabela_2_1 = pd.read_excel('./Dados_EB_orig.xls', sheet_name='Tabela 2.1', header=1) # aba tabela 2.1, primeira linha como cabeçalho
tabela_2_1

# Variáveis qualitativas (categóricas) descrevem atributos ou características não numéricas (ex: cor, sexo, escolaridade), dividindo-se em nominais (sem ordem) ou ordinais (com ordem). Variáveis quantitativas (numéricas) representam medidas ou contagens, sendo discretas (inteiros, ex: número de filhos) ou contínuas (decimais/intervalos, ex: altura, peso)

# Variaveis qualitativas nominais: Estadfo Civil, Região de Procedência
# Variaveis qualitativas ordinais: Grau de Instrução
# Variaveis quantitativas discretas: Número de Filhos, anos, meses
# variaveis quantitativas contínuas: Salario
```

```python
frequencia = tabela_2_1['Grau de Instrução'].value_counts() # conta a frequência de cada categoria na variável 'Grau de Instrução'
proporção = tabela_2_1['Grau de Instrução'].value_counts(normalize=True) # calcula a proporção de cada categoria
porcentagem = tabela_2_1['Grau de Instrução'].value_counts(normalize=True) * 100 # calcula a porcentagem de cada categoria
tabela_2_2 = pd.DataFrame({'Frequência': frequencia, 'Proporção': proporção, 'Porcentagem %': porcentagem}) # cria um DataFrame com frequência, proporção e porcentagem
tabela_2_2.loc['Total'] = tabela_2_2.sum().round(1) # adiciona uma linha 'Total' somando os valores numéricos. arrendonda para 1 casa decimal
tabela_2_2
```

```python
tabela_2_3 = pd.DataFrame({'Grau de Instrução': ['Fundamental', 'Médio', 'Superior'], 'Frequência': [1650, 1020, 1330], 'Porcentagem': [32.5, 51, 16.5]})
tabela_2_3.set_index('Grau de Instrução', inplace=True) # Grau de Instrução é indice principal
tabela_2_3.loc['Total'] = tabela_2_3.sum().round(1)
tabela_2_3
```

```python
import math

cortes = list(range(4, math.ceil(max(list(tabela_2_1['Salario (x Sal Min)']))) + 4, 4)) # criando lista de cortes de 4 em 4 a partir de 4 até o valor máximo arredondado para cima +4
tabela_prov = tabela_2_1.copy() # criando uma cópia temporária da tabela 2.1
tabela_prov['Faixa Salarial'] = pd.cut(tabela_2_1['Salario (x Sal Min)'], bins=cortes, right=False) # criando uma nova coluna 'Faixa Salarial' com intervalos definidos pelos cortes
frequencia = tabela_prov['Faixa Salarial'].value_counts().sort_index() # contando a frequência de cada faixa salarial e ordenando pelo índice (faixa salarial)
porcentagem = tabela_prov['Faixa Salarial'].value_counts(normalize=True).sort_index() * 100 # calculando a porcentagem de cada faixa salarial e ordenando pelo índice
tabela_2_4 = pd.DataFrame({'Frequência': frequencia, 'Porcentagem %': porcentagem}) # criando um DataFrame com frequência e porcentagem
del(tabela_prov) # deletando a tabela temporária
tabela_2_4.loc['Total'] = tabela_2_4.sum().round(1) # adicionando uma linha 'Total' somando os valores numéricos. arrendonda para 1 casa decimal
tabela_2_4

# frequencia faixa de salario, conseguir fazer aqui e no libreofiice, mas no excel nao consegui, pois teve problema em identificar como numerico e poder fazer o intervalo.
```
